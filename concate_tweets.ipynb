```python
import pandas as pd
import numpy as np
```

```python
dftwitter = pd.read_csv('data/tweets_withsentiment.csv')
dftwitter_sentiment = pd.read_csv('data/tweets_hashtag_withsentiment.csv')
```

```python
df = pd.concat([dftwitter, dftwitter_sentiment], axis=0)
```

```python
#Überprüfen ob die Daten korrekt gemerget wurden
dftwitter.shape[0] + dftwitter_sentiment.shape[0], df.shape[0]
```

```python
df_german = df[df['language_prob'] == 'de']
df_german = df_german[['date', 'content', 'language_prob', 'Positive', 'Negative', 'Neutral']]
df_german
```
