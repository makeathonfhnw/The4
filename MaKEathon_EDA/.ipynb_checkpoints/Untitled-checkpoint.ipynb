```python
# Module importieren
import pandas as pd # DataFrame
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle as pkl

# Machine Learning Module
```

```python
# Daten einlesen
app_store = pkl.load(open("../data/app_store.pkl", "rb"))
app_store = pd.json_normalize(app_store)

play_store = pkl.load(open("../data/play_store.pkl", "rb"))
play_store = pd.json_normalize(play_store)

twitter = pd.read_csv("../data/tweets_ats.csv")
```

```python
# App Store DataFrame bearbeiten

# Selektieren von interessanten Spalten
app_liste = ["title", "isEdited", "date", "review", "rating"]
app_store = app_store[app_liste]
app_store
```

```python
# app store
app_store["content"] = app_store["title"] + ": " + app_store["review"]
app_store
```

```python
# DataFrame ausgeben
app_store
```

```python
# Playstore DatenFarme bearbeiten
play_list = ['score', "content"]
play_store = play_store[play_list]
play_store = play_store.rename(columns={"score": "rating"})
play_store
```

```python
# Konkardieren
apple_playstore_df = pd.concat([app_store, play_store], axis = 0)
apple_playstore_df
```

```python
# Filtern nach jedem Rating
df_1 = apple_playstore_df[apple_playstore_df["rating"] == 1]
df_2 = apple_playstore_df[apple_playstore_df["rating"] == 2]
df_3 = apple_playstore_df[apple_playstore_df["rating"] == 3]
df_4 = apple_playstore_df[apple_playstore_df["rating"] == 4]
df_5 = apple_playstore_df[apple_playstore_df["rating"] == 5]
```
