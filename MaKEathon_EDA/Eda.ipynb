```python
# Module importieren
import pandas as pd # DataFrame
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle as pkl

# Machine Learning Module
```

```python
# Daten einlesen
app_store = pkl.load(open("../data/app_store.pkl", "rb"))
app_store = pd.json_normalize(app_store)

play_store = pkl.load(open("../data/play_store.pkl", "rb"))
play_store = pd.json_normalize(play_store)

twitter = pd.read_csv("../data/tweets_ats.csv")
```

```python
# App Store DataFrame bearbeiten

# Selektieren von interessanten Spalten
app_liste = ["title", "review", "rating"]
app_store = app_store[app_liste]
app_store
```

```python
# app store
app_store["content"] = app_store["title"] + ": " + app_store["review"]
app_store
```

```python
# select rating, content
app_store = app_store[["rating", "content"]]
app_store
```

```python
# Playstore DatenFarme bearbeiten
play_list = ['score', "content"]
play_store = play_store[play_list]
play_store = play_store.rename(columns={"score": "rating"})
play_store
```

```python
# Konkardieren
df = pd.concat([app_store, play_store], axis = 0)
df
```

```python
# filtern rating 1 
df_1 = df[df["rating"] == 1]
df_1
```
